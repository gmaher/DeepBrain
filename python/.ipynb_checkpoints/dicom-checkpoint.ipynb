```python
%matplotlib inline
import numpy as np
import SimpleITK as sitk
import matplotlib.pyplot as plt
```

```python
directory = './data/LED-FLAIR-DWI-masks/30023/'
directory_mask = directory+'FLAIR_MASK'
directory_flair = directory+'FLAIR'
directory_dwi = directory+'DWI'
```

```python
def dcmToNumpy(folder):
    reader = sitk.ImageSeriesReader()
    dicom_names = reader.GetGDCMSeriesFileNames(folder)
    reader.SetFileNames(dicom_names)
    image = reader.Execute()
    image_np = sitk.GetArrayFromImage(image)
    return image_np.copy()

image_mask_np = dcmToNumpy(directory_mask)
image_flair_np = dcmToNumpy(directory_flair)
image_dwi_np = dcmToNumpy(directory_dwi)
```

```python
# print dicom_names[0]
# print image.GetSize()
# print image
print image_mask_np.shape
print image_dwi_np.shape
print image_flair_np.shape
```

```python
slice = 15
plt.figure()
plt.imshow(image_mask_np[slice,:,:], cmap='gray')
plt.colorbar()
plt.show()

plt.figure()
plt.imshow(image_flair_np[slice,:,:], cmap='gray')
plt.colorbar()
plt.show()

plt.figure()
plt.imshow(image_dwi_np[slice,:,:], cmap='gray')
plt.colorbar()
plt.show()
```

```python
import os
folder = './data/LED-FLAIR-DWI-masks/'
dwis=[]
flairs=[]
masks=[]
for directory in os.listdir(folder):
    print directory
    dwi = dcmToNumpy(folder+'/'+directory+'/DWI')
    flair = dcmToNumpy(folder+'/'+directory+'/FLAIR')
    mask = dcmToNumpy(folder+'/'+directory+'/FLAIR_MASK')
    print "dwi shape: {}, flair shape {}, mask shape {}".format(
        dwi.shape,flair.shape,mask.shape)
    
    dwis.append(dwi)
    flairs.append(flair)
    masks.append(mask)
    
```

```python
for d in dwis:
    print 'mean = {}, std={}, max={}, min={}'.format(
        np.mean(d),np.std(d),np.amax(d),np.amin(d))
```

```python
for d in flairs:
    print 'mean = {}, std={}, max={}, min={}'.format(
        np.mean(d),np.std(d),np.amax(d),np.amin(d))
```

```python
for d in masks:
    print 'mean = {}, std={}, max={}, min={}'.format(
        np.mean(d),np.std(d),np.amax(d),np.amin(d))
```

```python
for i in range(len(dwis)):
    inds = masks[i] == 1
    #d = dwis[i][inds]
    f = flairs[i][inds]
    print 'flair, mean={}, std={}'.format(np.mean(f),np.std(f))
```

```python
for i in range(len(dwis)):
    inds = flairs[i] >= 100
    #d = dwis[i][inds]
    f = flairs[i][inds]
    print 'flair, mean={}, std={}'.format(np.mean(f),np.std(f))
```

```python
for i in range(len(dwis)):
    inds = masks[i] == 1
    #d = dwis[i][inds]
    f = flairs[i].copy()
    inds2 = f > 100
    f = (f-np.mean(f))/np.std(f)
    print 'flair, mean={}, std={}, nonzero, mean={}, std={}, stroke, mean={},std={}'.format(
        np.mean(f),np.std(f),np.mean(f[inds2]),np.std(f[inds2]),np.mean(f[inds]),np.std(f[inds]))
```

```python
import matplotlib.pyplot as plt

for i in range(len(dwis)):
    inds = masks[i] == 1
    #d = dwis[i][inds]
    f = flairs[i].copy()
    inds2 = f > 200
    f = (f-np.mean(f))/np.std(f)
    stroke = f[inds]
    nonzero = f[inds2]
    
    plt.figure()
    #plt.hist(np.ravel(f), bins=50,alpha=0.5,color='b',label='flair')
    plt.hist(np.ravel(stroke), bins=50,alpha=0.5,color='r',label='stroke')
    plt.hist(np.ravel(nonzero), bins=50,alpha=0.5,color='g',label='nonzero')
    plt.legend()
    plt.show()
```
