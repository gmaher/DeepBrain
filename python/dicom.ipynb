```python
%matplotlib inline
import numpy as np
import SimpleITK as sitk
import matplotlib.pyplot as plt
```

```python
directory = './data/LED-FLAIR-DWI-masks/30023/'
directory_mask = directory+'FLAIR_MASK'
directory_flair = directory+'FLAIR'
directory_dwi = directory+'DWI'
```

```python
reader = sitk.ImageSeriesReader()
dicom_names = reader.GetGDCMSeriesFileNames(directory_mask)
reader.SetFileNames(dicom_names)
image = reader.Execute()
image_mask_np = sitk.GetArrayFromImage(image)

dicom_names = reader.GetGDCMSeriesFileNames(directory_flair)
reader.SetFileNames(dicom_names)
image_flair = reader.Execute()
image_flair_np = sitk.GetArrayFromImage(image_flair)

dicom_names = reader.GetGDCMSeriesFileNames(directory_dwi)
reader.SetFileNames(dicom_names)
image_dwi = reader.Execute()
image_dwi_np = sitk.GetArrayFromImage(image_dwi)
```
